### Using Kaggle dataset create a logistic regression model for predict the fraud credit card.

## Imported Required Lib.

```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('creditcard.csv')
data.head()
```

```python
data.tail()
```

```python
data.shape
```

##  Data Preprocessing

```python
data.info()
```

```python
data.describe()
```

```python
data.duplicated().sum()
```

```python
data.isnull().sum()
```

```python
data.drop_duplicates(inplace = True)
```

```python
data.shape
```

```python
data['Class'].value_counts()
```

```python
data_fraud=data[data['Class']==1]
data_not_fraud=data[data['Class']==0]
```

```python
data_sample=data_not_fraud.sample(data_fraud.shape[0], random_state=70)
```

```python
final_data= pd.concat ([data_sample, data_fraud], axis=0).sample(frac=1, random_state=70). reset_index(). drop("index", axis=1)
final_data
```

##  Train Test Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
x=final_data.drop(columns='Class', axis=1)
y=final_data['Class']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y, test_size=0.2, random_state=50)
```

```python
print("X_tain",X_train.shape)
print("X_test",X_test.shape)
print("Y_tain",Y_train.shape)
print("Y_test",Y_test.shape)
```

## Build Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
Logi_reg= LogisticRegression()
```

```python
Logi_reg.fit(X_train,Y_train)
```

```python
y_pred=Logi_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score=accuracy_score(y_pred,Y_test)
```

```python
print("Accuracy Score :- ",score)
```
